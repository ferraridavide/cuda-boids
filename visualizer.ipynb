```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def visualize_boids_grid(path: str, title: str):
    df = pd.read_csv(path)
    plt.figure(figsize=(20, 16))
    unique_cells = df['cell_idx'].unique()
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_cells)))
    color_map = dict(zip(unique_cells, colors))
    point_colors = [color_map[cell] for cell in df['cell_idx']]

    scatter = plt.scatter(df['x'], df['y'], c=point_colors, s=5)
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title(title)
    plt.grid(True, alpha=0.3)

    cbar = plt.colorbar(scatter)
    cbar.set_label('Cell Index')

    plt.show()

def visualize_boids_index(path: str, title: str):
    df = pd.read_csv(path)

    colors = plt.cm.plasma(np.linspace(0, 1, len(df)))
    plt.figure(figsize=(20, 16))
    scatter = plt.scatter(df['x'], df['y'], c=colors, alpha=0.6, s=5)
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title(title)
    plt.grid(True, alpha=0.3)

    cbar = plt.colorbar(scatter)
    cbar.set_label('Point Index')

    plt.show()
```

```python
visualize_boids_index("1. Serial\\boids.csv", "Serial Boids")
visualize_boids_index("2. Parallel Naive\\boids_naive.csv", "Parallel Naive Boids")
visualize_boids_index("3. Parallel Grid\\boids_fast.csv", "Parallel Grid Boids")
visualize_boids_grid("3. Parallel Grid\\boids_fast.csv", "Parallel Grid Boids (Cell Index)")
```
